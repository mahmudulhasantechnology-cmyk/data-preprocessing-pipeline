```python
import pandas as pd

# Example using a valid, complete raw data URL
url = "https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv"
df = pd.read_csv(url)

```

```python
df['Age'].fillna(df['Age'].median())
```

```python
df.drop_duplicates(inplace=True)
```

```python
le = LabelEncoder()
df["Survived_LabelEncoded"] = le.fit_transform(df["Survived"])
print("Label Encoding (Survived) completed.")
```

```python
rank_order = [["Low", "Medium", "High"]]
oe = OrdinalEncoder(categories=rank_order)
df["Ticket_Class_OrdinalEncoded"] = oe.fit_transform(df[["Ticket_Class"]])
print("Ordinal Encoding (Ticket_Class) completed.")
```

```python
ohe = OneHotEncoder(sparse_output=False, drop="first")
sex_ohe = ohe.fit_transform(df[["Sex"]])
sex_ohe_df = pd.DataFrame(
    sex_ohe, columns=ohe.get_feature_names_out(["Sex"]), index=df.index
)


print("One-Hot Encoding (Sex) completed.\n")
```

```python
ss = StandardScaler()
df["Age_StandardScaled"] = ss.fit_transform(df[["Age"]])

# 2. MinMaxScaler (Scales values strictly between 0 and 1)
mms = MinMaxScaler()
df["Age_MinMaxScaled"] = mms.fit_transform(df[["Age"]])

# 3. RobustScaler (Uses Median and IQR; safe against extreme outliers)
rs = RobustScaler()
df["Fare_RobustScaled"] = rs.fit_transform(df[["Fare"]])

# 4. MaxAbsScaler (Scales data between -1 and 1 based on absolute maximums)
mas = MaxAbsScaler()
df["Fare_MaxAbsScaled"] = mas.fit_transform(df[["Fare"]])

# 5. Log Transform (Reduces skewness in highly asymmetrical features like Fare)
# np.log1p handles zero values safely by calculating log(x + 1)
df["Fare_LogTransformed"] = np.log1p(df["Fare"])

print("All scaling and transformations completed.\n")

```

```python
print("--- FINAL PROCESSED COLUMNS PREVIEW ---")
encoded_scaled_cols = [
    "Survived_LabelEncoded",
    "Ticket_Class_OrdinalEncoded",
    "Age_StandardScaled",
    "Age_MinMaxScaled",
    "Fare_RobustScaled",
    "Fare_MaxAbsScaled",
    "Fare_LogTransformed",
]
print(df[encoded_scaled_cols].head())
```
